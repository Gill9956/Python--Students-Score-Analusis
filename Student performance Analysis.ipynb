#Drop Unnamed column

```python
df= df.drop("Unnamed: 0",axis=1)
```

```python
print(df.head())
```

#change weekly study hours

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("5-Oct","5-10")
df.head()
```

#gender distribution

```python
plt.figure(figsize=(4,5))
ax=sns.countplot(data=df,x="Gender") #gives counted data
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

#this analysis show that number of females is more than that of males

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.title("Effect of parents education on their children's scores")
plt.show()
```

From above chart we have concluded that education of parents education have a great impact on childern marks

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot=True)
plt.title("Effect of parens martial status on their children's scores")
plt.show()
```

#From above chart we have concluded that the martial status of parents have no such impact on the kids score
